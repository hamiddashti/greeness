```python
import xarray as xr 
import matplotlib.pylab as plt
import numpy as np
import rioxarray 

dir = "/data/home/hamiddashti/hamid/nasa_above/greeness/"
```

```python
noaa = xr.open_dataarray("/data/home/hamiddashti/hamid/"
"nasa_above/greeness/data/processed_data/noaa_nc/lai_fapar/resampled/lai_growing.nc")
gimms = xr.open_dataarray("/data/home/hamiddashti/hamid/"
"nasa_above/greeness/data/processed_data/gimms/resampled/lai_growing_above.nc")
```

```python
noaa_mean = noaa.mean(["latitude","longitude"])
gimms_mean = gimms.mean(["lat","long"])

plt.figure(figsize=(12,8))
noaa_mean.plot(c="k",label="NOAA CDR LAI")
gimms_mean.plot(c="red",label="GIMMS LAI")
plt.title("Mean over the entire region",fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("")
plt.ylabel("LAI",fontsize=16)
plt.legend(fontsize=20)
plt.savefig(dir+"/working/Figures/Compare_Gimms_Noaa.png")
```

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
noaa.mean("time").plot()
plt.title("NOAA mean= "+str(noaa.mean().values),fontsize=18)
plt.subplot(1,2,2)
gimms.mean("time").plot()
plt.title("GIMMS mean= "+str(gimms.mean().values),fontsize=18)
plt.savefig(dir+"working/Figures/noaa_gimms_map.png")
```
