```r
library(doParallel)
library(foreach)

# Define the number of cores (adjust as needed)
num_cores <- 4

# Initialize parallel processing
cl <- makeCluster(num_cores)
registerDoParallel(cl)

# Define the dimensions of the matrix
matrix_rows <- 30
total_iterations <- 8  # Assuming you have defined 'total_iterations'

# Create a progress bar function (optional)
create_pb <- function(total) {
  pb <- txtProgressBar(min = 0, max = total, style = 3)
  on.exit(close(pb))
  return(pb)
}

# Define a function to fill a portion of the matrix and return results
calculate_recovery <- function(data, t) {
  # Perform your data processing here
  
  # For example:
  local_matrix <- matrix(0, nrow = 1, ncol = 5)
  y_result <- 123  # Replace with your actual y result
  
  # Return results as a named list
  return(list(local_matrix = local_matrix, y = y_result))
}

# Create the progress bar (optional)
# pb <- create_pb(total_iterations)

# Run the loop in parallel and accumulate results as a list
result_list <- foreach(i = 1:total_iterations, .combine = 'c') %dopar% {
  data <- ndvi_1994_data[i]
  data <- matrix(unlist(data), nrow = matrix_rows)
  
  # Perform the work and fill the local matrix, and get y
  results <- calculate_recovery(data, t)
  
  # Update the progress bar (optional)
  # setTxtProgressBar(pb, getTxtProgressBar(pb) + 1)
  
  return(results)
}

# Stop parallel processing
stopCluster(cl)

# Split 'result_list' into individual results
result_list <- split(result_list, rep(1:total_iterations, each = 1))

# Extract 'ci_list' and 'y_list' from the 'result_list'
ci_list <- lapply(result_list, function(result) result$local_matrix)
y_list <- lapply(result_list, function(result) result$y)
```
