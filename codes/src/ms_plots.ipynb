Calculating the contribution of lai trend due to lcc to the total trend compared to trend due to natural variability

```python
var_lcc_cont = lai_max_norm_trend_lcc_overlap.round(
    3) / lai_max_norm_trend_nv_overlap.round(3)  # contribution of LCC
var_lcc_cont = var_lcc_cont.where(
    np.isfinite(var_lcc_cont))  # There are division by zero
mask = np.isfinite(var_lcc_cont)
t = mask.sum().values  # Total number of pixels with detected LCC nad NV trends
con_lcc = (var_lcc_cont > 1).sum().values
con_nv = ((var_lcc_cont > 0) & (var_lcc_cont < 1)).sum().values
opp_con = (var_lcc_cont < 0).sum().values
print("LAI trend LCC > NV:" + str(np.round(100 * con_lcc / t)) + "%")
print("LAI trend NV > LCC:" + str(np.round(100 * con_nv / t)) + "%")
print("The estimated slope for LAI trend due to LCC is approximatley " +
      str(np.round(var_lcc_cont.where(var_lcc_cont > 1).weighted(weights).mean().values)) +
      " times larger than NV")
```

Same as above but for NDVI

```python
var_lcc_cont = ndvi_max_norm_trend_lcc.round(
    3) / ndvi_max_norm_trend_nv.round(3)  # contribution of LCC
var_lcc_cont = var_lcc_cont.where(
    np.isfinite(var_lcc_cont))  # There are division by zero
mask = np.isfinite(var_lcc_cont)
t = mask.sum().values  # Total number of pixels with detected LCC nad NV trends
con_lcc = (var_lcc_cont > 1).sum().values
con_nv = ((var_lcc_cont > 0) & (var_lcc_cont < 1)).sum().values
opp_sign = (var_lcc_cont < 0).sum().values
print("NDVI trend LCC > NV:" + str(np.round(100 * con_lcc / t)) + "%")
print("NDV trend NV > LCC:" + str(np.round(100 * con_nv / t)) + "%")
print("The estimated slope for NDVI trend due to LCC is approximatley " +
      str(np.round(var_lcc_cont.where(var_lcc_cont > 1).weighted(weights).mean().values)) +
      " times larger than NV")
```
